```python
from pandasai import SmartDataframe
from langchain_groq import ChatGroq
from google.colab import userdata
import pandas as pd
import numpy as np
```

```python
# supported models in Groq can be found at https://console.groq.com/docs/models
llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    api_key = userdata.get('GROQ_API_KEY'),
    )
```

```python
url = "https://drive.google.com/file/d/1H5lV6GG0CyzuThhwBaNrNrmS1wT2sDF7/view?usp=sharing"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
nls97 = pd.read_csv(path)
nls97.head()
```

```python
nls97.set_index("personid", inplace=True)
nls97.describe()
```

```python
nls97.shape
```

```python
nls97sdf = SmartDataframe(nls97, config={"llm": llm, "temperature": 0, "verbose": False, "output_type": "plot", "custom_whitelisted_dependencies":["scikit-learn","statsmodels"], "enable_cache": True, "use_error_correction_framework": False})
```

```python
if nls97sdf is not None:
    print(nls97sdf.head())
else:
    print("SmartDataframe is None. Check your setup.")
```

```python
nls97.isna().sum().sum()
```

```python
nls97['hdegnum'] = nls97.highestdegree.str[0:1].astype('category')
```

```python
nls97['parentincome'] = nls97.parentincome.replace(list(range(-5,0)),np.nan)
```

```python
wagedatalist = ['wageincome20','weeksworked20','parentincome','hdegnum']
```

```python
wagedata = nls97[wagedatalist]
```

```python
wagedatasdf = SmartDataframe(wagedata, config={"llm": llm, "temperature": 0, "verbose": False, "output_type": "plot", "custom_whitelisted_dependencies":["scikit-learn","statsmodels"], "enable_cache": True, "use_error_correction_framework": False})
```

```python
wagedatasdf.chat("Show the counts, means, and standard deviations of columns on a table")
```

```python
wagedatasdf.chat("Impute missings based on most frequent value")
```

```python
df = wagedatasdf.chat("convert to pandas dataframe")
```
